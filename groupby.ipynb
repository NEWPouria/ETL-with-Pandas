```python
import pandas as pd
import numpy as np
from tabulate import tabulate

# ایجاد داده‌های نمونه از فروشگاه
data = {
    'شهر': ['تهران', 'تهران', 'مشهد', 'مشهد', 'اصفهان', 'اصفهان', 'تهران', 'مشهد'],
    'محصول': ['لپ‌تاپ', 'موبایل', 'لپ‌تاپ', 'تبلت', 'موبایل', 'تبلت', 'تبلت', 'موبایل'],
    'فروشنده': ['علی', 'رضا', 'مریم', 'محمد', 'فاطمه', 'حسین', 'نازنین', 'کاظم'],
    'مبلغ_فروش': [1500, 800, 1400, 600, 900, 550, 700, 850],
    'تعداد': [3, 5, 2, 4, 6, 3, 2, 7],
    'ماه': ['فروردین', 'فروردین', 'اردیبهشت', 'فروردین', 'اردیبهشت', 'فروردین', 'اردیبهشت', 'فروردین']
}

df = pd.DataFrame(data)

print("📊 داده‌های فروشگاه:")
df
```

```python
# میانگین مبلغ فروش به تفکیک شهر

df.groupby('شهر')['مبلغ_فروش'].mean()
df.groupby('شهر').agg({'مبلغ_فروش': 'mean'})
```

```python
# مجموع تعداد فروش به تفکیک شهر و محصول

df.groupby(['شهر','محصول']).agg({'تعداد':'sum'})
```

```python
# میانگین مبلغ فروش و مجموع تعداد به تفکیک شهر

df.groupby('شهر').agg({'مبلغ_فروش':'mean','تعداد':'sum'})
```

```python
# برای هر شهر: بیشترین مبلغ فروش، کمترین تعداد، و تعداد فروش‌های انجام شده

df.groupby('شهر').agg({
    'مبلغ_فروش': 'max',      # بیشترین مبلغ فروش
    'تعداد': 'min',          # کمترین تعداد
    'فروشنده': 'count'       # تعداد فروش‌های انجام شده
})
```

```python
# میانگین تعداد فروش به تفکیک محصول

df.groupby('محصول').agg({'تعداد':'mean'})
```

```python
# بیشترین مبلغ فروش به تفکیک شهر و ماه

df.groupby(['شهر','ماه']).agg({'مبلغ_فروش':'max'})
```

```python
# برای هر محصول: میانگین مبلغ فروش و مجموع تعداد

df.groupby('محصول').agg({'تعداد':'sum','مبلغ_فروش':'mean'})
```

```python
# برای هر شهر: تعداد محصولات مختلف فروخته شده، میانگین تعداد و مبلغ فروش

df.groupby('شهر').agg({
    'محصول': 'nunique',      # تعداد محصولات مختلف
    'تعداد': 'mean',         # میانگین تعداد
    'مبلغ_فروش': 'mean'      # میانگین مبلغ فروش
})


# nunique(): تعداد مقادیر منحصر به فرد (بدون تکرار)
# count(): تعداد کل رکوردها (همه مقادیر حتی تکراری)
```

```python
# لیست فروشنده های تمام شهر ها رو بگیرم؟
# تمام شهر ها و فروشنده هاشون

df.groupby('شهر')['فروشنده'].unique()
df.groupby('شهر').agg({'فروشنده': 'unique'})
```

```python
# تعداد فروشندگان مختلف در هر شهر

df.groupby('شهر').agg({'فروشنده': 'nunique'})
df.groupby('شهر').agg({'فروشنده': 'count'})
```

```python
# میانگین مبلغ فروش به تفکیک محصول و ماه

df.groupby(['محصول', 'ماه']).agg({'مبلغ_فروش':'mean'})
df.groupby([ 'ماه' ,'محصول']).agg({'مبلغ_فروش':'mean'})
```

```python
# برای هر ماه: مجموع تعداد، بیشترین مبلغ فروش و تعداد شهرهای مختلف

df.groupby('ماه').agg({'تعداد':'sum','مبلغ_فروش':'max','شهر':'nunique'})
```

```python
# برای هر محصول: اولین و آخرین مبلغ فروش، و اختلاف بین بیشترین و کمترین مبلغ فروش

df.groupby('محصول').agg(
    اولین_فروش=('مبلغ_فروش', 'first'),
    آخرین_فروش=('مبلغ_فروش', 'last'), 
    بیشترین_فروش=('مبلغ_فروش', 'max'),
    کمترین_فروش=('مبلغ_فروش', 'min'),
    اختلاف_فروش=('مبلغ_فروش', lambda x: x.max() - x.min())
)
```

```python
import pandas as pd
bios=pd.read_excel("C:/Users/LENOVO/Downloads/olympics-data.xlsx")
```

```python
# **** تبدیل به تاریخ ****
bios['born_date']=pd.to_datetime(bios['born_date'])
# ***** reset_index()=Generate a new DataFrame or Series with the index reset. *****
# ***** صرفا برای شمارش از ستون name استفاده میکنه *****
result = bios.groupby(bios['born_date'].dt.year)['name'].count().reset_index().sort_values(by='name', ascending=False)
result
```

```python
bios['born_date']=pd.to_datetime(bios['born_date'])
bios['month_date'] = bios['born_date'].dt.month
bios['year_date'] = bios['born_date'].dt.year
bios.groupby(['year_date', 'month_date'])['name'].count().reset_index().sort_values('name',ascending=False)
```
