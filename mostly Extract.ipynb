```python
# فیلتر کردن با دو شرط
bios.loc[(bios['height_cm']>215) & (bios['born_country']=='USA'),['name','height_cm','born_country']]

# فیلتر کردن براساس اسم دقیق
bios.loc[(bios['name'].str.contains('hamed|pouria',case=False)) & (bios['born_country'].str.lower()=='iri') & (bios['height_cm']>180)]
```
